```python
import pandas as pd
from matplotlib import pyplot as plt
import locale
import os
from pandas import read_csv
from datetime import datetime, timezone
```

```python
dir_path2 = "./DataSet_FBK/StChiara"
res = []
df = []
for path in os.listdir(dir_path2):
    if os.path.isfile(os.path.join(dir_path2, path)):
        res.append(path)
for file in res:
    df.append(read_csv(dir_path2+"/"+file, sep = ",", index_col=False ))
import pytz
for i in range(6):
    print(df[i])
    df[i] = df[i].drop("Unnamed: 0",axis = 1)      
    try:
        df[i].ts = pd.to_datetime(df[i].ts)
        df[i].ts = df[i].ts.dt.tz_localize('UTC')

    except Exception as e: 
        print(e)
        print(i)
        df[i].ts = pd.to_datetime(df[i].ts.str.slice(start = 0, stop = 24), format = "%a %b %d %Y %H:%M:%S")
        df[i].ts = df[i].ts.dt.tz_localize('Europe/London', ambiguous="NaT", nonexistent = "NaT")
    print(df[i])
```

```python
change = {}
for i in range(1,9):
    change["Rs" + str(i)] = "S" + str(i) + "_R2"
    change["Rh" + str(i)] = "S" + str(i) + "_R1"
    change["Vh" + str(i)] = "S" + str(i) + "_Voltage"
change["BME"] = "IAQ"
    
```

```python
df[0] = df[0].rename(columns=change)
```

```python
df[0].columns
```

```python
df_Tot = pd.concat(df)
```

```python
df_Tot = df_Tot.drop(["CFG", "TH", "H", "G",  "CO2", "VOC", "IAC_comp"], axis = 1)
```

```python
df_Tot = df_Tot.dropna()
df_Tot
```

```python
df_Tot.ts = df_Tot.ts.apply(lambda x : x.astimezone('UTC'))
```

```python
df_Tot = df_Tot.set_index("ts")
```

```python
df_Tot = df_Tot.resample("1T").mean()
df_Tot = df_Tot.dropna()
df_Tot.to_csv(dir_path2 + ".csv")
```
